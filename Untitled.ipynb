```python
from pathlib import Path
from typing import Tuple

import cv2
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pydicom
from tqdm.notebook import tqdm
```

```python
def load_and_prepare_labels(labels_csv_path: Path) -> pd.DataFrame:
    """
    Зчитує CSV із мітками та повертає очищену таблицю без дублікатів.

    Таблиця приводиться до двох колонок:
    - ``patientId`` — ідентифікатор знімка;
    - ``Target`` — мітка класу (0 — норма, 1 — пневмонія).

    Параметри:
        labels_csv_path (Path): Шлях до CSV-файла з мітками.

    Повертає:
        pandas.DataFrame: Таблиця з колонками ``patientId`` і ``Target``,
        у якій для кожного пацієнта залишено лише одну унікальну мітку.
    """
    if not labels_csv_path.exists():
        raise FileNotFoundError(f"Файл з мітками не знайдено: {labels_csv_path}")

    labels = pd.read_csv(labels_csv_path)
    labels = labels[["patientId", "Target"]]
    labels = labels.drop_duplicates("patientId")

    return labels
```

```python
def plot_sample_grid(
    labels: pd.DataFrame,
    images_root: Path,
    n_rows: int = 3,
    n_cols: int = 3,
) -> None:
    """
    Виводить сітку з прикладами DICOM-знімків та відповідними мітками.
 
    Параметри:
        labels (pandas.DataFrame):
            Таблиця з колонками ``patientId`` і ``Target``.
        images_root (Path):
            Коренева папка з DICOM-файлами (stage_2_train_images).
        n_rows (int, optional):
            Кількість рядків у сітці. За замовчуванням 3.
        n_cols (int, optional):
            Кількість стовпців у сітці. За замовчуванням 3.

    """
    fig, axis = plt.subplots(n_rows, n_cols, figsize=(3 * n_cols, 3 * n_rows))
    c = 0

    for i in range(n_rows):
        for j in range(n_cols):
            patient_id = labels.patientId.iloc[c]
            dcm_path = (images_root / patient_id).with_suffix(".dcm")

            if not dcm_path.exists():
                raise FileNotFoundError(f"DICOM-файл не знайдено: {dcm_path}")

            dcm = pydicom.dcmread(dcm_path).pixel_array
            label = labels.Target.iloc[c]

            axis[i][j].imshow(dcm, cmap="bone")
            axis[i][j].set_title(str(label))
            axis[i][j].axis("off")
            c += 1

    plt.tight_layout()
    plt.show()
```

```python
def process_and_save_images(
    labels: pd.DataFrame,
    images_root: Path,
    save_root: Path,
    image_size: Tuple[int, int] = (224, 224),
    train_size: int = 24_000,
) -> Tuple[float, float]:
    """
    Обробляє DICOM-зображення, зберігає їх у форматі .npy та рахує mean/std
    для train-частини.

    Параметри:
        labels (pandas.DataFrame):
            Таблиця з колонками ``patientId`` і ``Target``.
        images_root (Path):
            Коренева папка з DICOM-зображеннями.
        save_root (Path):
            Коренева папка, куди будуть збережені оброблені .npy файли.
        image_size (tuple[int, int], optional):
            Розмір вихідного зображення (width, height) для моделі.
            За замовчуванням (224, 224).
        train_size (int, optional):
            Кількість перших прикладів, які вважаються train-частиною.

    Повертає:
        tuple[float, float]:
            Пара (mean, std), обчислена по train-частині датасету.
    """
    save_root.mkdir(parents=True, exist_ok=True)

    sums = 0.0
    sums_squared = 0.0

    for idx, patient_id in enumerate(tqdm(labels.patientId, desc="Обробка DICOM-зображень")):
        dcm_path = (images_root / patient_id).with_suffix(".dcm")

        if not dcm_path.exists():
            raise FileNotFoundError(f"DICOM-файл не знайдено: {dcm_path}")

        dcm = pydicom.dcmread(dcm_path).pixel_array.astype(np.float32) / 255.0

        dcm_array = cv2.resize(dcm, image_size).astype(np.float16)

        label = int(labels.Target.iloc[idx])
        train_or_val = "train" if idx < train_size else "val"

        current_save_path = save_root / train_or_val / str(label)
        current_save_path.mkdir(parents=True, exist_ok=True)

        np.save(current_save_path / patient_id, dcm_array)

        # накопичуємо статистики лише по train-частині
        if train_or_val == "train":
            normalizer = dcm_array.shape[0] * dcm_array.shape[1]
            pixel_sum = np.sum(dcm_array) / normalizer
            pixel_sum_sq = np.power(dcm_array, 2).sum() / normalizer

            sums += pixel_sum
            sums_squared += pixel_sum_sq

    mean = sums / train_size
    std = np.sqrt(sums_squared / train_size - (mean ** 2))

    return float(mean), float(std)
```

```python
if __name__ == "__main__":
    LABELS_CSV_PATH = Path(
        "/Users/romyna_m/Documents/pneumonia_class/"
        "rsna-pneumonia-detection-challenge/stage_2_train_labels.csv"
    )
    ROOT_PATH = Path(
        "/Users/romyna_m/Documents/pneumonia_class/"
        "rsna-pneumonia-detection-challenge/stage_2_train_images/"
    )
    SAVE_PATH = Path("/Users/romyna_m/Documents/pneumonia_class/Processed/")

    # Зчитуємо та готуємо мітки
    labels_df = load_and_prepare_labels(LABELS_CSV_PATH)

    # Виводимо кілька прикладів знімків
    plot_sample_grid(labels_df, ROOT_PATH, n_rows=3, n_cols=3)

    # Обробляємо всі знімки та рахуємо mean/std
    mean, std = process_and_save_images(
        labels=labels_df,
        images_root=ROOT_PATH,
        save_root=SAVE_PATH,
        image_size=(224, 224),
        train_size=24_000,
    )

    print(f"Середнє значення: {mean:.6f}, Стандартне відхилення: {std:.6f}")
```
