```python
%matplotlib inline
import torch
import torchvision
from torchvision import transforms
import torchmetrics
import pytorch_lightning as pl
from pytorch_lightning.callbacks import ModelCheckpoint, EarlyStopping
from pytorch_lightning.loggers import TensorBoardLogger
from tqdm.notebook import tqdm
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_file(path):
    return np.load(path).astype(np.float32)
```

```python
train_transforms = transforms.Compose([
                                    transforms.ToTensor(),
                                    transforms.Normalize(0.49, 0.248),
                                    transforms.RandomAffine(
                                        degrees=(-5, 5), translate=(0, 0.05), scale=(0.9, 1.1)),
                                        transforms.RandomResizedCrop((224, 224), scale=(0.35, 1))

])

val_transforms = transforms.Compose([
                                    transforms.ToTensor(),
                                    transforms.Normalize([0.49], [0.248]),
])
```

```python
from pathlib import Path

train_root = Path("/Users/romyna_m/Documents/pneumonia_class/Processed/train")

val_root = Path("/Users/romyna_m/Documents/pneumonia_class/Processed/val")

train_dataset = torchvision.datasets.DatasetFolder(
    train_root,
    loader=load_file,
    extensions=("npy",),
    transform=train_transforms
)

val_dataset = torchvision.datasets.DatasetFolder(
    val_root,
    loader=load_file,
    extensions=("npy",),
    transform=val_transforms
)
```

```python
from torch.utils.data import WeightedRandomSampler
import numpy as np

targets = np.array(train_dataset.targets)

class_counts = np.bincount(targets)

class_weights = 1.0 / class_counts

sample_weights = class_weights[targets]

sampler = WeightedRandomSampler(
    weights=sample_weights,
    num_samples=len(sample_weights),
    replacement=True
)
```

```python
fig, axis = plt.subplots(2, 2, figsize=(9, 9))
for i in range(2):
    for j in range(2):
        random_index = np.random.randint(0, 20000)
        x_ray, label = train_dataset[random_index]
        axis[i][j].imshow(x_ray[0], cmap="bone")
        axis[i][j].set_title(f"Label:{label}")
```

```python
batch_size = 64
num_workers = 4

train_loader = torch.utils.data.DataLoader(
    train_dataset, batch_size=batch_size, sampler=sampler, num_workers=num_workers, persistent_workers=True)

val_loader = torch.utils.data.DataLoader(
    val_dataset, batch_size=batch_size, num_workers=num_workers, shuffle=False, persistent_workers=True)


print(f"There are {len(train_dataset)} train images and {len(val_dataset)} val images")
```

```python
np.unique(train_dataset.targets, return_counts=True), np.unique(val_dataset.targets, return_counts=True)
```

```python
class PneumoniaModel(pl.LightningModule):
    def __init__(self, weight=1):
        super().__init__()

        self.model = torchvision.models.resnet18()
        self.model.conv1 = torch.nn.Conv2d(1, 64, kernel_size=(7, 7), stride=(2, 2), padding=(3, 3), bias=False)
        self.model.fc = torch.nn.Linear(in_features=512, out_features=1)

        self.optimizer = torch.optim.Adam(self.model.parameters(), lr=1e-4)
        self.loss_fn = torch.nn.BCEWithLogitsLoss(pos_weight=torch.tensor([weight]))


        self.train_acc = torchmetrics.Accuracy(task="binary")
        self.val_acc = torchmetrics.Accuracy(task="binary")

    def forward(self, data):
        return self.model(data)

    def training_step(self, batch, batch_idx):
        x_ray, label = batch
        label = label.float()
        pred = self(x_ray)[:, 0]

        loss = self.loss_fn(pred, label)

        probs = torch.sigmoid(pred)
        self.train_acc(probs, label.int())

        self.log("Train Loss", loss, on_step=True, on_epoch=True)
        self.log("Train Acc", self.train_acc, on_step=False, on_epoch=True, prog_bar=True)
        return loss

    def validation_step(self, batch, batch_idx):
        x_ray, label = batch
        label = label.float()
        pred = self(x_ray)[:, 0]

        loss = self.loss_fn(pred, label)

        probs = torch.sigmoid(pred)
        self.val_acc(probs, label.int())

        self.log("Val Loss", loss, on_step=False, on_epoch=True)
        self.log("Val Acc", self.val_acc, on_step=False, on_epoch=True, prog_bar=True)
        return loss

    def configure_optimizers(self):
        return [self.optimizer]
```

```python
model = PneumoniaModel()
```

```python
checkpoint_callback = ModelCheckpoint(
    monitor='Val Acc',      # monitors what you already log
    save_top_k=10,
    mode='max'
)

early_stop_callback = EarlyStopping(
    monitor='Val Loss',     # EXACTLY the same string as in self.log(...)
    mode='min',             # lower loss is better
    patience=5,             # stop if no improvement for 5 epochs
    min_delta=0.0           # minimal change to count as "improved"
)
```

Trainer documentation: https://pytorch-lightning.readthedocs.io/en/latest/common/trainer.html

```python
gpus = 1
trainer = pl.Trainer(
    accelerator="gpu",
    devices=1,
    logger=TensorBoardLogger(save_dir="./logs"),
    log_every_n_steps=1,
    callbacks=[checkpoint_callback, early_stop_callback],
    max_epochs=60,   # or 80  a higher UPPER bound is fine now
)
```

```python
trainer.fit(model, train_loader, val_loader)
```

```python
best_path = checkpoint_callback.best_model_path
print("Best model checkpoint:", best_path)
```

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")

model = PneumoniaModel.load_from_checkpoint(best_path, strict=False)
model.eval()
model.to(device);
```

```python
preds = []
labels = []

with torch.no_grad():
    for data, label in tqdm(val_dataset):
        data = data.to(device).float().unsqueeze(0)
        pred = torch.sigmoid(model(data)[0].cpu())
        preds.append(pred)
        labels.append(label)
preds = torch.tensor(preds)
labels = torch.tensor(labels).int()
```

```python
acc = torchmetrics.Accuracy(task="binary")(preds, labels)
precision = torchmetrics.Precision(task="binary")(preds, labels)
recall = torchmetrics.Recall(task="binary")(preds, labels)

metrics = {
    ".": acc.item(),        # Accuracy
    ".": precision.item(), # Precision
    ".": recall.item()       # Recall
}

plt.figure(figsize=(6, 4))
plt.bar(metrics.keys(), metrics.values())
plt.ylim(0, 1)
plt.title(" ")
plt.ylabel("")
plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.show()
```

```python
cm = torchmetrics.ConfusionMatrix(task="binary")(preds, labels)
cm_np = cm.numpy()

fig, ax = plt.subplots(figsize=(4, 4))
im = ax.imshow(cm_np, interpolation="nearest")
ax.set_title(" ")
ax.set_xticks([0, 1])
ax.set_yticks([0, 1])
ax.set_xticklabels([" ", ""])
ax.set_yticklabels([" ", ""])
plt.xlabel("")
plt.ylabel("")

for i in range(2):
    for j in range(2):
        ax.text(j, i, cm_np[i, j], ha="center", va="center")

plt.colorbar(im, ax=ax)
plt.tight_layout()
plt.show()
```
